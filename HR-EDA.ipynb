## Total Number of Jobs : 1128

```python
min_exp = []
max_exp = []

for i in df['Experience']:
    try:
        min_exp.append(i.split(' ')[0])
    except:
        min_exp.append(' ')
        
    try:
        max_exp.append(i.split(' ')[2])
    except:
        max_exp.append('-')

df['Minimum Experience'] = min_exp
df['Maximum Experience'] = max_exp
```

```python
df
```

```python
class DataAnalysis:
    def details(self, df):
        print('Number of (Rows, Columns): ',df.shape,'\n')
        print('Columns in Dataframe: \n',df.columns,'\n')
        print('\nDateframe Details:')
        print(df.info(),'\n')
        print('Statistical Summary: \n',df.describe(),'\n')
        print('Null Value Count: \n',df.isnull().sum(),'\n')
        
    def column_insights(self, df):
        for i in df.columns:
            print('\nColumn Name -----> ',i)
            print('\nNumber of Unique Values: ',df[i].nunique())
            print('\nCount of Unique Values:\n',df[i].value_counts())
    
    def drop_duplicates(self, df):
        if df.duplicated().sum() > 0:
            print('Dataframe contains duplicate Values', df.duplicated().sum(),'\n')
            df.drop_duplicates(inplace=True,ignore_index=True)
            print('After droping duplicates\n')
            self.details(df)
            
        else:
            print("Dataframe doesn't contain any duplicate value")
```

```python
da = DataAnalysis()
```

```python
da.drop_duplicates(df)
```

```python
da.details(df)
```

```python
# There are 13 Null Job Titles, hence below code is to remove that particular tuple
df.dropna(subset=['Job Title'], inplace=True)
```
