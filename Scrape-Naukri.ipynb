```python
from selenium.webdriver.common.by import By
from selenium import webdriver
import logging as console
import pandas as pd
import time
import csv
```

```python
console.basicConfig(filename= "Search-Jobs.log", level= console.DEBUG, format= '%(asctime)s%(message)s')

class SearchJobDetails:
    
    def get_driver(self):
        console.info('\nPreparing Driver')
        DRIVER_PATH = 'Enter your Chrome Driver path'
        driver = webdriver.Chrome(executable_path = DRIVER_PATH)
        return driver
    
    def get_job_details(self, skills, location):
        console.info('\nGetting Job Details..')
        driver = self.get_driver()
        
        driver.get("https://www.naukri.com/")
        driver.implicitly_wait(5)
        
        # enter skills 
        skill = driver.find_element(By.XPATH,'//*[@id="root"]/div[6]/div/div/div[1]/div/div/div/input')
        skill.send_keys(skills)
        
        # enter location
        loc = driver.find_element(By.XPATH,'//*[@id="root"]/div[6]/div/div/div[5]/div/div/div/input')
        loc.send_keys(location)
        driver.implicitly_wait(5)
        
        # click submit button
        search_btn = driver.find_element(By.CLASS_NAME,'qsbSubmit')
        search_btn.click()
        time.sleep(2)
        
        filename = skills+' '+location + '.csv'
                
        headers = ['Job Title','Company','Location','Job Highlights','Full Job Description','Reviews',\
                    'Job Type','Qualification','Experience','Salary','Posted on','Required Skills',\
                    'Application Link','Search Term']
        
        # csv to write all the job details
        with open(filename, 'w', newline="", encoding='UTF-8', errors='ignore') as csv_file:
            writer = csv.DictWriter(csv_file, fieldnames=headers, extrasaction='ignore')
            writer.writeheader()
            
            while True:
                
                try:
                    nextPagebtn = driver.find_element(By.CSS_SELECTOR,'#root > div.search-result-container > \
                    div > section.listContainer.fleft > div.mt-64.mb-60 > div > \
                    a.fright.fs14.btn-secondary.br2')
                    
                except:
                    print('Fetched all jobs')
                    break

                job_list = driver.find_element(By.CLASS_NAME,'list').find_elements(By.TAG_NAME,'article')
                
                
                for job in job_list:
                    try:
                        driver.switch_to.window(driver.window_handles[0])
                        
                        while True:
                            try:
                                # first click job card
                                job.click()     
                                driver.implicitly_wait(5)
                                # switch driver to opened tab
                                driver.switch_to.window(driver.window_handles[1]) 
                                break
                            except:
                                pass

                        try:
                            title = driver.find_element(By.CLASS_NAME,'jd-header-title').text
                        except:
                            title = ''

                        try:
                            company = driver.find_element(By.CSS_SELECTOR,'#root > main > div.jd-container > \
                            div.leftSec > section.jd-header > div.top > div.jd-top-head > div > a.pad-rt-8').text
                        except:
                            company = ''

                        try:
                            location = driver.find_element(By.CSS_SELECTOR,'#root > main > div.jd-container > \
                            div.leftSec > section.jd-header > div.top > div.left > div.loc > span').text
                        except:
                            location = ''

                        try:
                            jd = driver.find_element(By.CSS_SELECTOR,'#root > main > div.jd-container > \
                            div.leftSec > section.job-desc > div.dang-inner-html > p:nth-child(1)').text
                        except:
                            jd = ''

                        try:
                            full_jd = driver.find_element(By.CSS_SELECTOR,'#root > main > div.jd-container \
                            > div.leftSec > section.job-desc > div.dang-inner-html').text 
                            
                            full_jd = full_jd.replace('\n',' ')
                        except:
                            full_jd = ''
                            

                        time.sleep(2)

                        try:
                            reviews = driver.find_element(By.CSS_SELECTOR,'#root > main > div.jd-container > \
                            div.leftSec > section.jd-header > div.top > div.jd-top-head > div > a:nth-child(2) > \
                            span.amb-reviews.pad-rt-4').text.split(' ')[0].split('(')[1]
                        except:
                            reviews = ''


                        other_details = driver.find_element(By.CSS_SELECTOR,'#root > main > div.jd-container > \
                        div.leftSec > section.job-desc > div.other-details').text

                        with open('otherDetails.txt','w',encoding='UTF-8', errors='ignore') as jobdetails:
                            text = other_details
                            jobdetails.write(text)
                            jobdetails.close()

                        with open('otherDetails.txt',encoding='UTF-8', errors='ignore') as lines:
                            f = lines.readlines()
                            jobtype = ''

                            for i in range(len(f)):
                                try:
                                    if 'Employment' in f[i]:
                                        jobtype = ''.join(f[i].split('Type')[1].replace('\n',''))
                                except:
                                    jobtype = 'Not Mentioned'



                        qual = []
                        qualification = driver.find_element(By.CSS_SELECTOR,'#root > main > div.jd-container \
                        > div.leftSec > section.job-desc > div.education').text

                        try:
                            qual.append(qualification.splitlines()[1])
                            try:
                                qual.append(qualification.splitlines()[2])
                            except:
                                pass
                        except:
                            qual.append('Not Mentioned')

                        q = ', '.join(qual)


                        try:
                            experience = driver.find_element(By.CSS_SELECTOR,'#root > main > div.jd-container > \
                            div.leftSec > section.jd-header > div.top > div.left > div.exp > span').text
                        except:
                            experience = 'Not Mentioned'


                        salary = driver.find_element(By.CSS_SELECTOR,'#root > main > div.jd-container > \
                        div.leftSec > section.jd-header > div.top > div.left > div.salary > span').text

                        try:
                            posted_on = driver.find_element(By.CSS_SELECTOR,'#root > main > div.jd-container > \
                            div.leftSec > section.jd-header > div.bottom > div.jd-stats > span:nth-child(1) > \
                            span').text
                        except:
                            posted_on = ''


                        skills_list = []
                        try:
                            skills_loc = driver.find_element(By.XPATH,'//*[@id="root"]/main/div[2]/div[2]\
                            /section[2]/div[4]/div[2]').find_elements(By.TAG_NAME,'a')

                            for skill in skills_loc:
                                skills_list.append()
                        except:
                            pass

                        try:
                            skills_loc = driver.find_element(By.XPATH,\
                                        '//*[@id="root"]/main/div[2]/div[2]/section[2]/div[4]/div[3]')\
                                        .find_elements(By.TAG_NAME,'a')

                            for skill in skills_loc:
                                skills_list.append(skill.text)
                        except:
                            pass


                        s = ', '.join(skills_list)

                        try:
                            link = driver.current_url
                        except:
                            link = ''


                        # appending all the collected data to the dictionary
                        job_details_dict = {
                                    'Job Title': title,
                                    'Company': company,
                                    'Location': location,
                                    'Job Highlights': jd,
                                    'Full Job Description': full_jd,
                                    'Reviews': reviews,
                                    'Job Type': jobtype,
                                    'Qualification': q,
                                    'Experience': experience,
                                    'Salary': salary,
                                    'Posted on': posted_on,
                                    'Required Skills': s,
                                    'Application Link': link,
                                    'Search Term': skills
                                    }

                        try:
                            writer.writerow(job_details_dict)
                        except Exception as e:
                            print(e)

                        job_details_dict.clear()

                        # close current window
                        driver.switch_to.window(driver.window_handles[1])
                        driver.close()
                        
                        
                    except:
                        driver.switch_to.window(driver.window_handles[1])
                        driver.close()
                
                # switching to the parent window
                driver.switch_to.window(driver.window_handles[0])
                    
                nextPagebtn.click()
                driver.implicitly_wait(2)
                time.sleep(5)
                
        return 'Done fetching '+skill
        
    def data_science_jobs(self):
        console.info('\nInside data_science_jobs()')
        self.get_job_details(skills='data scientist', location='india')
        return 
        
    def data_analyst_jobs(self):
        console.info('\nInside data_analyst_jobs()')
        self.get_job_details(skills='data analyst', location='india')
        return 
    
    def data_engineer_jobs(self):
        console.info('\nInside data_engineer_jobs()')
        self.get_job_details(skills='data engineer', location='india')
        return 
    
    def machine_learning_engineer_jobs(self):
        console.info('\nInside machine_learning_engineer_jobs()')
        self.get_job_details(skills='machine learning engineer', location='india')
        return 
```
