```python
import os
import pandas as pd
```

```python
# enter path to CSV files
path = os.getcwd()
```

```python
# List with all CSV files
csv_files = [path+'/'+file for file in os.listdir(path) if file.endswith('.csv')]
```

```python
csv_data = []
```

```python
for file in csv_files:
    csv_data.append(pd.read_csv(file))
```

```python
csv_merged = pd.concat(csv_data, ignore_index=True)
```

```python
csv_merged.to_csv(path+'/merged-job-details.csv',index=False)
```
